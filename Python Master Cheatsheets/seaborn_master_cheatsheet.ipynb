```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(42)
print('Seaborn version:', sns.__version__)

# ── Reusable sample datasets ────────────────────────────────────
df = pd.DataFrame({
    'A':    np.random.randint(1, 100, 100),
    'B':    np.random.randint(1, 100, 100),
    'C':    np.random.choice(['X','Y','Z'], 100),
    'D':    np.random.choice(['M','F'], 100),
    'score':np.random.normal(70, 15, 100).clip(0, 100)
})

# Built-in datasets (always available)
tips     = sns.load_dataset('tips')
iris     = sns.load_dataset('iris')
titanic  = sns.load_dataset('titanic')
penguins = sns.load_dataset('penguins')
flights  = sns.load_dataset('flights')
```

---
## 1. THEMES & STYLES

```python
# ── Style (background) ──────────────────────────────────────────
# Options: darkgrid, whitegrid, dark, white, ticks
sns.set_style('darkgrid')
sns.set_style('whitegrid')
sns.set_style('white')
sns.set_style('ticks')

# Style with overrides
sns.set_style('whitegrid', {'axes.facecolor': '#f8f8f8',
                             'grid.color': 'white'})

# Context (scale for different outputs)
# Options: paper, notebook (default), talk, poster
sns.set_context('notebook')    # default
sns.set_context('talk')        # larger text for presentations
sns.set_context('poster')      # biggest — conference posters
sns.set_context('paper')       # smallest

# Context with custom scale
sns.set_context('notebook', font_scale=1.4, rc={'lines.linewidth': 2})

# Palettes
sns.set_palette('Set2')        # qualitative
sns.set_palette('Blues')       # sequential
sns.set_palette('RdBu')        # diverging
sns.set_palette('husl', 5)     # n colors from husl
sns.set_palette('colorblind')  # accessible

# All-in-one
sns.set_theme(style='whitegrid', palette='Set2', context='notebook', font_scale=1.2)

# Reset
sns.reset_defaults()
sns.reset_orig()               # matplotlib defaults

# Despine (remove top/right border)
sns.despine()                  # removes top + right
sns.despine(left=True)         # remove left too
sns.despine(offset=10)         # offset spines from axes
sns.despine(trim=True)         # trim to data range
```

---
## 2. COLOR PALETTES

```python
# View palette
sns.palplot(sns.color_palette('Set2'))
plt.show()

# Qualitative (categories)
sns.color_palette('Set1')          # 9 distinct colors
sns.color_palette('Set2')          # muted
sns.color_palette('Set3')          # pastel
sns.color_palette('tab10')         # matplotlib default
sns.color_palette('Paired')        # paired colors
sns.color_palette('colorblind')    # accessible

# Sequential (low → high)
sns.color_palette('Blues', 8)
sns.color_palette('viridis', 8)
sns.color_palette('rocket')        # seaborn native
sns.color_palette('mako')
sns.color_palette('flare')
sns.color_palette('crest')

# Diverging (low ← mid → high)
sns.color_palette('RdBu', 11)
sns.color_palette('coolwarm', 11)
sns.color_palette('vlag')
sns.color_palette('icefire')

# Custom palette
custom = sns.color_palette(['#e74c3c','#2ecc71','#3498db'])
sns.set_palette(custom)

# husl / hls (evenly spaced hues)
sns.color_palette('husl', 6)
sns.color_palette('hls',  6)

# Light/dark palette from one color
sns.light_palette('navy', n_colors=6)
sns.dark_palette('purple', n_colors=6)
sns.diverging_palette(220, 20, n=11)  # hue1, hue2, n colors

# Use as context
with sns.color_palette('rocket'):
    sns.lineplot(data=df, x='A', y='B', hue='C')
    plt.show()
```

---
## 3. RELATIONAL PLOTS — lineplot / scatterplot / relplot

```python
# ── scatterplot ─────────────────────────────────────────────────
sns.scatterplot(data=tips, x='total_bill', y='tip')
plt.show()

# hue / style / size — three extra dimensions
sns.scatterplot(data=tips, x='total_bill', y='tip',
                hue='sex',      # color by column
                style='smoker', # marker style by column
                size='size',    # marker size by column
                palette='Set1', alpha=0.8)
plt.show()

# ── lineplot ────────────────────────────────────────────────────
fmri = sns.load_dataset('fmri')
sns.lineplot(data=fmri, x='timepoint', y='signal',
             hue='region', style='event',
             errorbar='sd')      # show std dev band
plt.show()

# No confidence interval
sns.lineplot(data=fmri, x='timepoint', y='signal',
             hue='region', errorbar=None)
plt.show()

# errorbar options: 'ci', 'pi', 'se', 'sd', None

# ── relplot — figure-level (returns FacetGrid) ──────────────────
g = sns.relplot(data=tips, x='total_bill', y='tip',
                col='time',      # separate column per time
                hue='smoker',
                style='smoker',
                kind='scatter')  # or 'line'
g.set_axis_labels('Total Bill ($)', 'Tip ($)')
g.set_titles(col_template='{col_name} meals')
plt.show()
```

---
## 4. DISTRIBUTION PLOTS

```python
# ── histplot ────────────────────────────────────────────────────
sns.histplot(data=tips, x='total_bill')
plt.show()

# KDE overlay
sns.histplot(data=tips, x='total_bill', kde=True, bins=20)
plt.show()

# Grouped histogram
sns.histplot(data=tips, x='total_bill', hue='sex', multiple='stack')
# multiple options: 'layer', 'dodge', 'stack', 'fill'
plt.show()

# 2D histogram
sns.histplot(data=tips, x='total_bill', y='tip', bins=20, cmap='Blues')
plt.show()

# Normalized (stat options: count, frequency, density, probability)
sns.histplot(data=tips, x='total_bill', stat='density', kde=True)
plt.show()

# ── kdeplot ─────────────────────────────────────────────────────
sns.kdeplot(data=tips, x='total_bill')
plt.show()

# Grouped
sns.kdeplot(data=tips, x='total_bill', hue='sex', fill=True, alpha=0.3)
plt.show()

# 2D KDE (bivariate)
sns.kdeplot(data=tips, x='total_bill', y='tip', fill=True, cmap='Blues')
plt.show()

# Bandwidth control
sns.kdeplot(data=tips, x='total_bill', bw_adjust=0.5)  # narrower
sns.kdeplot(data=tips, x='total_bill', bw_adjust=2.0)  # smoother
plt.show()

# ── ecdfplot (empirical cumulative distribution) ─────────────────
sns.ecdfplot(data=tips, x='total_bill', hue='sex')
plt.show()

# ── rugplot ─────────────────────────────────────────────────────
sns.kdeplot(data=tips, x='total_bill')
sns.rugplot(data=tips, x='total_bill', height=0.05, alpha=0.4)
plt.show()

# ── displot — figure-level ──────────────────────────────────────
sns.displot(data=tips, x='total_bill', col='time',
            hue='sex', kde=True, kind='hist')   # hist, kde, ecdf
plt.show()
```

---
## 5. CATEGORICAL PLOTS

```python
# ── stripplot (scatter for categories) ──────────────────────────
sns.stripplot(data=tips, x='day', y='total_bill', hue='sex',
              jitter=True, alpha=0.7, dodge=True)
plt.show()

# ── swarmplot (non-overlapping points) ──────────────────────────
sns.swarmplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)
plt.show()

# ── boxplot ─────────────────────────────────────────────────────
sns.boxplot(data=tips, x='day', y='total_bill')
plt.show()

# With hue (grouped)
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex', palette='Set2')
plt.show()

# Box + strip overlay
sns.boxplot(data=tips, x='day', y='total_bill', palette='Set3')
sns.stripplot(data=tips, x='day', y='total_bill',
              color='black', alpha=0.4, size=3)
plt.show()

# ── violinplot ──────────────────────────────────────────────────
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex',
               split=True,     # mirror halves per hue
               inner='quart',  # inner: quartile, box, stick, None
               palette='muted')
plt.show()

# ── boxenplot (letter-value plot — better for large data) ────────
sns.boxenplot(data=tips, x='day', y='total_bill', hue='sex')
plt.show()

# ── pointplot (mean + CI per category) ──────────────────────────
sns.pointplot(data=tips, x='day', y='total_bill', hue='sex',
              dodge=True, capsize=0.1, errorbar='sd')
plt.show()

# ── barplot ─────────────────────────────────────────────────────
sns.barplot(data=tips, x='day', y='total_bill', hue='sex',
            errorbar='sd', capsize=0.1, palette='Set1')
plt.show()

# ── countplot ───────────────────────────────────────────────────
sns.countplot(data=tips, x='day', hue='sex', palette='Set2')
plt.show()

# Horizontal count
sns.countplot(data=tips, y='day', order=tips['day'].value_counts().index)
plt.show()

# ── catplot — figure-level for all categorical plots ─────────────
g = sns.catplot(data=tips, x='day', y='total_bill',
                hue='sex', col='time',
                kind='box',    # strip swarm box violin boxen point bar count
                height=4, aspect=0.8)
g.set_axis_labels('Day', 'Total Bill')
plt.show()
```

---
## 6. REGRESSION PLOTS

```python
# ── regplot (axes-level) ─────────────────────────────────────────
sns.regplot(data=tips, x='total_bill', y='tip')
plt.show()

# No CI band
sns.regplot(data=tips, x='total_bill', y='tip', ci=None)
plt.show()

# Polynomial regression
sns.regplot(data=tips, x='total_bill', y='tip', order=2)
plt.show()

# Logistic regression (for binary y)
sns.regplot(data=titanic, x='age', y='survived', logistic=True,
            scatter_kws={'alpha': 0.2})
plt.show()

# Lowess (non-parametric smooth)
sns.regplot(data=tips, x='total_bill', y='tip', lowess=True, ci=None)
plt.show()

# Robust regression (reduces outlier influence)
sns.regplot(data=tips, x='total_bill', y='tip', robust=True)
plt.show()

# Residual plot
sns.residplot(data=tips, x='total_bill', y='tip', lowess=True)
plt.show()

# ── lmplot — figure-level (supports facets + hue) ────────────────
sns.lmplot(data=tips, x='total_bill', y='tip',
           hue='sex', col='time', row='smoker',
           height=4, aspect=0.8, ci=95)
plt.show()
```

## 7. MATRIX PLOTS — heatmap / clustermap

```python
# ── heatmap ─────────────────────────────────────────────────────
corr = tips.select_dtypes('number').corr()

sns.heatmap(corr)
plt.show()

# Annotated with values
sns.heatmap(corr, annot=True, fmt='.2f', linewidths=0.5)
plt.show()

# Custom colormap & range
sns.heatmap(corr, annot=True, cmap='RdBu',
            vmin=-1, vmax=1, center=0,
            square=True, linecolor='white')
plt.show()

# Mask upper triangle (show only lower)
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, annot=True, mask=mask, cmap='coolwarm',
            vmin=-1, vmax=1, fmt='.2f')
plt.show()

# Pivot table heatmap
pivot = flights.pivot(index='month', columns='year', values='passengers')
sns.heatmap(pivot, annot=True, fmt='d', cmap='YlOrRd')
plt.title('Passengers by Month & Year')
plt.show()

# ── clustermap (hierarchical clustering + heatmap) ───────────────
sns.clustermap(pivot, cmap='viridis', standard_scale=1,
               figsize=(10,8), linewidths=0.3)
plt.show()

# No reordering (just heatmap look with clustermap API)
sns.clustermap(pivot, row_cluster=False, col_cluster=False, cmap='Blues')
plt.show()
```

## 8. PAIRPLOT & PAIRGRID

```python
# ── pairplot ─────────────────────────────────────────────────────
sns.pairplot(iris)
plt.show()

# Color by category
sns.pairplot(iris, hue='species')
plt.show()

# Custom diagonal (kde or hist)
sns.pairplot(iris, hue='species',
             diag_kind='kde',          # 'hist' or 'kde'
             plot_kws={'alpha': 0.5})
plt.show()

# Subset columns
sns.pairplot(iris, vars=['sepal_length','petal_length','petal_width'],
             hue='species', kind='reg')  # kind: scatter, reg, kde, hist
plt.show()

# Corner (only lower triangle)
sns.pairplot(iris, corner=True, hue='species')
plt.show()

# ── PairGrid (full manual control) ──────────────────────────────
g = sns.PairGrid(iris, hue='species')
g.map_upper(sns.scatterplot, alpha=0.6)
g.map_lower(sns.kdeplot)
g.map_diag(sns.histplot, kde=True)
g.add_legend()
plt.show()
```

## 9. FACETGRID

```python
# ── FacetGrid ───────────────────────────────────────────────────
g = sns.FacetGrid(tips, col='time', row='smoker',
                  hue='sex', height=3, aspect=1.2,
                  margin_titles=True)
g.map(sns.scatterplot, 'total_bill', 'tip', alpha=0.6)
g.add_legend()
g.set_axis_labels('Total Bill', 'Tip')
g.set_titles(col_template='{col_name}', row_template='{row_name}')
plt.show()

# map_dataframe (pass DataFrame + column names)
g = sns.FacetGrid(tips, col='time', height=4)
g.map_dataframe(sns.histplot, x='total_bill', kde=True)
plt.show()

# Custom function in map
def custom_scatter(x, y, **kwargs):
    plt.scatter(x, y, **kwargs)
    plt.axhline(y.mean(), color='red', ls='--')

g = sns.FacetGrid(tips, col='day', height=3)
g.map(custom_scatter, 'total_bill', 'tip', alpha=0.5)
plt.show()

# Share axes or not
g = sns.FacetGrid(tips, col='time', sharex=False, sharey=False)
g.map(sns.histplot, 'total_bill')
plt.show()
```

## 10. JOINTPLOT & JOINTGRID

```python
# ── jointplot ───────────────────────────────────────────────────
sns.jointplot(data=tips, x='total_bill', y='tip')
plt.show()

# kind options: scatter, kde, hist, hex, reg, resid
sns.jointplot(data=tips, x='total_bill', y='tip', kind='kde',
              fill=True, cmap='Blues')
plt.show()

sns.jointplot(data=tips, x='total_bill', y='tip', kind='hex',
              gridsize=20)
plt.show()

sns.jointplot(data=tips, x='total_bill', y='tip', kind='reg')
plt.show()

# With hue
sns.jointplot(data=tips, x='total_bill', y='tip',
              hue='sex', kind='scatter')
plt.show()

# Custom marginal kws
sns.jointplot(data=tips, x='total_bill', y='tip',
              marginal_kws=dict(bins=15, fill=True),
              joint_kws=dict(alpha=0.4))
plt.show()

# ── JointGrid (full manual control) ─────────────────────────────
g = sns.JointGrid(data=tips, x='total_bill', y='tip', height=6)
g.plot_joint(sns.scatterplot, hue=tips['sex'], alpha=0.6)
g.plot_marginals(sns.histplot, kde=True)
plt.show()
```

## 11. CUSTOMIZATION — TITLES, LABELS, LEGENDS

```python
fig, ax = plt.subplots(figsize=(8,5))

sns.barplot(data=tips, x='day', y='total_bill', hue='sex',
            palette='Set2', ax=ax)

# Title & labels
ax.set_title('Total Bill by Day & Sex', fontsize=14, fontweight='bold', pad=15)
ax.set_xlabel('Day of Week', fontsize=12)
ax.set_ylabel('Total Bill ($)', fontsize=12)

# Tick customization
ax.set_xticklabels(ax.get_xticklabels(), rotation=30, ha='right')
ax.tick_params(axis='both', labelsize=10)

# Legend
ax.legend(title='Sex', title_fontsize=11, fontsize=10,
          loc='upper right', framealpha=0.7)

# Grid
ax.grid(axis='y', linestyle='--', alpha=0.5)

sns.despine(ax=ax)
plt.tight_layout()
plt.show()

# ── Figure-level plot customization ─────────────────────────────
g = sns.catplot(data=tips, x='day', y='total_bill',
                hue='sex', col='time', kind='box')

g.set_axis_labels('Day', 'Total Bill ($)')
g.set_titles(col_template='Meal: {col_name}')
g.set(ylim=(0, 60))
g.despine(left=True)
g.fig.suptitle('Bill Distribution by Day', y=1.03, fontsize=14)
g.add_legend(title='Gender')
plt.show()
```

## 13. AXES-LEVEL vs FIGURE-LEVEL

```python
# ── Axes-level: plug into existing subplot ───────────────────────
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.boxplot(data=tips, x='day', y='total_bill', ax=axes[0])
axes[0].set_title('Box Plot')

sns.scatterplot(data=tips, x='total_bill', y='tip',
                hue='sex', ax=axes[1])
axes[1].set_title('Scatter')

plt.tight_layout()
plt.show()

# All axes-level functions (ax= param)
# sns.scatterplot()  sns.lineplot()   sns.barplot()
# sns.boxplot()      sns.violinplot() sns.histplot()
# sns.kdeplot()      sns.regplot()    sns.heatmap()
# sns.countplot()    sns.stripplot()  sns.swarmplot()

# ── Figure-level: manage their own figure ────────────────────────
# Returns FacetGrid object, NOT axes
# Use g.fig / g.axes / g.set_*()

# relplot()  → scatterplot / lineplot + facets
# displot()  → histplot / kdeplot / ecdfplot + facets
# catplot()  → all categorical plots + facets
# lmplot()   → regplot + facets
# pairplot() → PairGrid shortcut
# jointplot()→ JointGrid shortcut

g = sns.displot(data=tips, x='total_bill', col='time', hue='sex', kind='kde')
g.fig.set_size_inches(10, 4)
plt.show()
```

## 14. OVERLAYING PLOTS

```python
# Box + strip (raw points on top)
fig, ax = plt.subplots(figsize=(8,5))
sns.boxplot(data=tips, x='day', y='total_bill', palette='pastel', ax=ax)
sns.stripplot(data=tips, x='day', y='total_bill',
              color='black', size=3, alpha=0.4, jitter=True, ax=ax)
plt.show()

# Violin + swarm
fig, ax = plt.subplots(figsize=(8,5))
sns.violinplot(data=tips, x='day', y='total_bill',
               inner=None, palette='muted', ax=ax)
sns.swarmplot(data=tips, x='day', y='total_bill',
              color='white', size=3, ax=ax)
plt.show()

# KDE + rug
fig, ax = plt.subplots(figsize=(8,4))
sns.kdeplot(data=tips, x='total_bill', hue='sex', fill=True, alpha=0.3, ax=ax)
sns.rugplot(data=tips, x='total_bill', hue='sex', height=0.05, ax=ax)
plt.show()

# Scatter + reg line
fig, ax = plt.subplots(figsize=(8,5))
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex',
                alpha=0.6, ax=ax)
sns.regplot(data=tips, x='total_bill', y='tip',
            scatter=False, color='black', ax=ax)  # scatter=False: line only
plt.show()
```

## 15. OBJECT INTERFACE (Seaborn v0.12+)

```python
# Modern declarative API — composes plot layers
from seaborn import objects as so

# Basic
(so.Plot(tips, x='total_bill', y='tip')
 .add(so.Dot())
 .show())

# With color grouping
(so.Plot(tips, x='total_bill', y='tip', color='sex')
 .add(so.Dot(alpha=0.6))
 .show())

# Regression line layer
(so.Plot(tips, x='total_bill', y='tip', color='sex')
 .add(so.Dot(alpha=0.4))
 .add(so.Line(), so.PolyFit(order=1))
 .show())

# Faceting
(so.Plot(tips, x='total_bill', y='tip', color='sex')
 .add(so.Dot(alpha=0.5))
 .facet(col='time')
 .show())

# Bar summary
(so.Plot(tips, x='day', y='total_bill')
 .add(so.Bar(), so.Agg('mean'))
 .add(so.Range(), so.Est(errorbar='sd'))
 .show())
```
