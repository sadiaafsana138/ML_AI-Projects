## 4. SELECTION — loc / iloc / at / iat

```python
df = pd.DataFrame({'A': [10,20,30], 'B': [40,50,60], 'C': [70,80,90]},
                   index=['x','y','z'])

# Column select
df['A']                    # Series
df[['A', 'B']]             # DataFrame

# loc — label based
df.loc['x']                # row by label
df.loc['x':'y']            # row slice (inclusive)
df.loc['x', 'A']           # single value
df.loc[['x','z'], ['A','C']]  # multi rows & cols
df.loc[:, 'A':'B']         # column slice

# iloc — position based
df.iloc[0]                 # first row
df.iloc[0:2]               # rows 0,1
df.iloc[0, 1]              # row 0, col 1
df.iloc[[0,2], [0,2]]      # fancy
df.iloc[:, 1:]             # all rows, cols from 1

# at / iat — fastest scalar access
df.at['x', 'A']            # label based single value
df.iat[0, 0]               # position based single value

# Boolean indexing in loc
df.loc[df['A'] > 10]
df.loc[df['A'] > 10, ['B', 'C']]
```

## 5. FILTERING & QUERY

```python
df = pd.DataFrame({'name':['Alice','Bob','Carol','Dave'],
                   'dept':['IT','HR','IT','HR'],
                   'salary':[100,200,150,250],
                   'age':[25,30,22,35]})

# Boolean conditions
df[df['salary'] > 150]
df[(df['salary'] > 100) & (df['dept'] == 'IT')]
df[(df['salary'] < 150) | (df['age'] > 30)]
df[~(df['dept'] == 'IT')]          # NOT IT

# isin
df[df['dept'].isin(['IT', 'Finance'])]
df[~df['dept'].isin(['HR'])]

# between
df[df['salary'].between(100, 200)]  # inclusive by default

# str filtering
df[df['name'].str.startswith('A')]
df[df['name'].str.contains('o', case=False)]

# query — cleaner syntax for large chains
df.query('salary > 150 and dept == "IT"')
df.query('salary > @threshold')    # use Python variable with @

# filter (filter columns/rows by label)
df.filter(items=['name','salary'])         # keep these cols
df.filter(like='sal', axis=1)             # cols containing 'sal'
df.filter(regex='^s', axis=1)             # cols starting with 's'
```

## 6. ADD / MODIFY / DELETE COLUMNS & ROWS

```python
df = pd.DataFrame({'A': [1,2,3], 'B': [4,5,6]})

# Add column
df['C'] = df['A'] + df['B']
df['D'] = 100                        # broadcast scalar

# assign (chainable, returns new df)
df = df.assign(E=lambda x: x['A'] * 10,
               F=lambda x: x['E'] + x['B'])  # can reference just-created col

# insert at position
df.insert(loc=1, column='Z', value=99)

# Modify column
df['A'] = df['A'] * 2
df.loc[0, 'A'] = 999               # modify single cell

# Add row
new_row = pd.DataFrame({'A': [10], 'B': [20]})
df = pd.concat([df, new_row], ignore_index=True)

# Drop columns
df.drop('D', axis=1, inplace=True)
df.drop(['E','F'], axis=1, inplace=True)

# Drop rows
df.drop(0, axis=0)                  # by label
df.drop([0, 1], axis=0)

# pop (remove and return column)
col = df.pop('Z')

# del
del df['C']
```

## 8. DTYPE CONVERSION

```python
df = pd.DataFrame({'A': ['1','2','3'], 'B': [1.1, 2.2, 3.3],
                   'C': ['2024-01-01','2024-02-01','2024-03-01']})

# Convert single column
df['A'] = df['A'].astype(int)
df['B'] = df['B'].astype(np.float32)

# Convert multiple
df = df.astype({'A': float, 'B': int})

# Safe numeric conversion (errors='coerce' → NaN on failure)
df['A'] = pd.to_numeric(df['A'], errors='coerce')

# Datetime
df['C'] = pd.to_datetime(df['C'])
df['C'] = pd.to_datetime(df['C'], format='%Y-%m-%d')

# Categorical (saves memory for low-cardinality cols)
df['dept'] = pd.Categorical(['IT','HR','IT'])
df['dept'] = df['dept'].astype('category')

# infer_objects — auto convert object cols
df = df.infer_objects()

# convert_dtypes — best nullable dtypes
df = df.convert_dtypes()
```

## 9. NULL / MISSING VALUE HANDLING

```python
df = pd.DataFrame({'A': [1, np.nan, 3, np.nan],
                   'B': [4, 5, np.nan, np.nan],
                   'C': ['x', 'y', None, 'z']})

# Detect
df.isnull()                         # True where NaN
df.notnull()                        # True where NOT NaN
df.isnull().sum()                   # count NaNs per col
df.isnull().sum().sum()             # total NaNs
df.isnull().mean() * 100            # % missing per col

# Drop
df.dropna()                         # drop rows with ANY NaN
df.dropna(axis=1)                   # drop cols with ANY NaN
df.dropna(how='all')                # drop rows where ALL are NaN
df.dropna(subset=['A', 'B'])        # only check these cols
df.dropna(thresh=2)                 # keep rows with >= 2 non-NaN

# Fill
df.fillna(0)                        # fill all with 0
df.fillna({'A': 0, 'B': 99})        # per-column fill
df['A'].fillna(df['A'].mean())      # fill with mean
df.fillna(method='ffill')           # forward fill (deprecated in newer pandas)
df.ffill()                          # forward fill (new API)
df.bfill()                          # backward fill
df.fillna(method='ffill', limit=1)  # fill max 1 consecutive

# Interpolate
df['A'].interpolate()               # linear interpolation
df['A'].interpolate(method='polynomial', order=2)

# Replace (not just NaN)
df.replace(99, np.nan)              # replace value
df.replace({'A': {0: np.nan}})      # column-specific
df.replace([1, 2], [10, 20])        # list replace
```

## 11. GROUPBY — FULL GUIDE

```python
df = pd.DataFrame({
    'dept':  ['IT','HR','IT','HR','IT'],
    'gender':['M','F','F','M','M'],
    'salary':[100,200,150,250,120],
    'age':   [25,30,22,35,28]
})

grp = df.groupby('dept')

# Built-in aggregations
grp['salary'].mean()
grp['salary'].sum()
grp['salary'].min()
grp['salary'].max()
grp['salary'].count()
grp['salary'].std()
grp['salary'].median()

# Multiple columns
grp[['salary','age']].mean()

# Multiple aggregations with agg
grp['salary'].agg(['mean','std','min','max'])
grp.agg({'salary': ['mean','max'], 'age': 'min'})

# Named aggregations (clean output)
grp.agg(
    avg_salary=('salary', 'mean'),
    max_salary=('salary', 'max'),
    headcount=('salary', 'count')
)

# Custom function
grp['salary'].agg(lambda x: x.max() - x.min())  # range

# Multi-key groupby
df.groupby(['dept','gender'])['salary'].mean()

# transform — returns same shape as original (no collapse)
df['salary_mean'] = df.groupby('dept')['salary'].transform('mean')
df['salary_zscore'] = df.groupby('dept')['salary'].transform(
    lambda x: (x - x.mean()) / x.std()
)

# filter — keep groups satisfying condition
df.groupby('dept').filter(lambda x: x['salary'].mean() > 150)

# apply — most flexible (any function)
def top_earner(g):
    return g.nlargest(1, 'salary')
df.groupby('dept').apply(top_earner)

# size vs count
grp.size()    # total rows including NaN
grp.count()   # non-null count per col

# as_index=False (returns DataFrame, not index)
df.groupby('dept', as_index=False)['salary'].mean()
```

## 12. MERGE / JOIN

```python
df1 = pd.DataFrame({'id':[1,2,3], 'name':['A','B','C']})
df2 = pd.DataFrame({'id':[1,2,4], 'salary':[100,200,300]})

# merge (SQL-style)
pd.merge(df1, df2, on='id')                  # inner join (default)
pd.merge(df1, df2, on='id', how='left')      # left join
pd.merge(df1, df2, on='id', how='right')     # right join
pd.merge(df1, df2, on='id', how='outer')     # full outer join
pd.merge(df1, df2, on='id', how='cross')     # cartesian product

# Different key names
df3 = pd.DataFrame({'emp_id':[1,2], 'dept':['IT','HR']})
pd.merge(df1, df3, left_on='id', right_on='emp_id')

# Merge on index
pd.merge(df1, df2, left_index=True, right_index=True)

# Suffix for overlapping columns
pd.merge(df1, df2, on='id', suffixes=('_left','_right'))

# indicator — adds _merge column showing source
pd.merge(df1, df2, on='id', how='outer', indicator=True)

# validate — check join type
pd.merge(df1, df2, on='id', validate='1:1')   # assert one-to-one

# join (index-based shortcut)
df1.set_index('id').join(df2.set_index('id'), how='left')

# merge_asof — nearest key merge (sorted)
left  = pd.DataFrame({'time':[1,5,10], 'val':[10,20,30]})
right = pd.DataFrame({'time':[2,6,9],  'ref':[100,200,300]})
pd.merge_asof(left, right, on='time')   # backward match
```

## 14. APPLY / MAP / APPLYMAP

```python
df = pd.DataFrame({'A':[1,2,3], 'B':[4,5,6]})

# apply on Series (element-wise)
df['A'].apply(lambda x: x**2)
df['A'].apply(np.sqrt)

# apply on DataFrame (column-wise by default)
df.apply(lambda col: col.max() - col.min())  # axis=0
df.apply(lambda row: row.sum(), axis=1)       # row-wise

# apply with result_type
df.apply(lambda row: pd.Series({'sum': row.sum(), 'mean': row.mean()}), axis=1)

# map (Series only) — replaces values
s = pd.Series(['cat','dog','cat'])
s.map({'cat': 1, 'dog': 0})                  # dict mapping
s.map(lambda x: x.upper())                   # function mapping

# map on DataFrame column
df['A'].map(lambda x: f'val_{x}')

# applymap / map (DataFrame element-wise)
# pandas >= 2.1: use df.map()
# pandas < 2.1:  use df.applymap()
df.map(lambda x: x * 10)   # newer pandas

# pipe — chain custom functions
def add_total(df):
    df['total'] = df.sum(axis=1)
    return df

df.pipe(add_total).pipe(lambda df: df[df['total'] > 5])
```

## 16. DATETIME OPERATIONS (dt accessor)

```python
df = pd.DataFrame({'date': pd.date_range('2024-01-15', periods=6, freq='M')})
df['ts'] = pd.to_datetime(['2024-01-15 10:30:00',
                            '2024-03-22 14:45:00',
                            '2024-06-01 08:00:00',
                            '2024-09-10 23:59:00',
                            '2024-11-05 05:15:00',
                            '2024-12-31 12:00:00'])

ts = df['ts']

# Extract components
ts.dt.year
ts.dt.month
ts.dt.day
ts.dt.hour
ts.dt.minute
ts.dt.second
ts.dt.weekday           # Mon=0, Sun=6
ts.dt.day_of_week
ts.dt.day_name()        # 'Monday', 'Tuesday'...
ts.dt.month_name()
ts.dt.quarter
ts.dt.day_of_year
ts.dt.week              # week number
ts.dt.isocalendar()     # ISO year, week, day

# Boolean properties
ts.dt.is_month_start
ts.dt.is_month_end
ts.dt.is_year_start
ts.dt.is_year_end
ts.dt.is_leap_year

# Rounding
ts.dt.floor('H')        # floor to hour
ts.dt.ceil('H')         # ceil to hour
ts.dt.round('H')        # round to nearest hour

# Timedelta / differences
df['diff'] = ts - ts.shift(1)     # time between rows
df['days'] = df['diff'].dt.days
df['seconds'] = df['diff'].dt.total_seconds()

# Offset / shift dates
ts + pd.Timedelta(days=7)
ts + pd.DateOffset(months=1)

# Date range
pd.date_range('2024-01-01', periods=12, freq='MS')   # month start
pd.date_range('2024-01-01', '2024-12-31', freq='W')  # weekly
pd.bdate_range('2024-01-01', periods=5)               # business days

# Resample (time-series groupby)
df_ts = df.set_index('ts')
df_ts.resample('M').mean()    # monthly mean
df_ts.resample('Q').sum()     # quarterly sum
df_ts.resample('W').last()    # last value per week

# Format to string
ts.dt.strftime('%Y-%m-%d')
ts.dt.strftime('%d/%m/%Y %H:%M')
```

## 18. PIVOT TABLE / CROSSTAB / MELT / WIDE_TO_LONG

```python
df = pd.DataFrame({
    'dept':   ['IT','HR','IT','HR','IT'],
    'gender': ['M','F','F','M','M'],
    'year':   [2022,2022,2023,2023,2023],
    'salary': [100,200,150,250,120]
})

# Pivot table
pd.pivot_table(df, values='salary', index='dept', columns='year',
               aggfunc='mean', fill_value=0, margins=True)

# pivot (no aggregation — must be unique)
df2 = pd.DataFrame({'date':['2024-01','2024-01','2024-02'],
                    'item':['A','B','A'], 'val':[10,20,30]})
df2.pivot(index='date', columns='item', values='val')

# crosstab — frequency table
pd.crosstab(df['dept'], df['gender'])
pd.crosstab(df['dept'], df['gender'], values=df['salary'], aggfunc='mean')
pd.crosstab(df['dept'], df['gender'], normalize='index')  # row %

# melt — wide to long
wide = pd.DataFrame({'id':[1,2], 'jan':[ 10,20], 'feb':[30,40], 'mar':[50,60]})
pd.melt(wide, id_vars='id', var_name='month', value_name='sales')

# stack / unstack
df3 = df.set_index(['dept','gender'])['salary']
df3.unstack()            # rows → columns
df3.unstack().stack()    # back to long

# wide_to_long
df4 = pd.DataFrame({'id':[1,2], 'sal_2022':[100,200], 'sal_2023':[150,250]})
pd.wide_to_long(df4, stubnames='sal', i='id', j='year', sep='_')
```

## 19. WINDOW FUNCTIONS (Rolling / Expanding / EWM)

```python
df = pd.DataFrame({'val': [10, 20, 15, 30, 25, 40, 35]})

# Rolling (fixed window)
df['roll_mean']  = df['val'].rolling(3).mean()
df['roll_sum']   = df['val'].rolling(3).sum()
df['roll_std']   = df['val'].rolling(3).std()
df['roll_max']   = df['val'].rolling(3).max()
df['roll_min']   = df['val'].rolling(3).min()

# min_periods (compute even with fewer values at start)
df['val'].rolling(3, min_periods=1).mean()

# center=True (window centered on current row)
df['val'].rolling(3, center=True).mean()

# Expanding (all previous values)
df['exp_mean'] = df['val'].expanding().mean()    # cumulative mean
df['exp_max']  = df['val'].expanding().max()     # running max

# EWM (Exponential Weighted Moving Average)
df['ewm_mean'] = df['val'].ewm(span=3).mean()   # span-based
df['ewm_mean'] = df['val'].ewm(alpha=0.5).mean()  # alpha-based

# shift / lag
df['lag1']  = df['val'].shift(1)    # lag 1 (previous row)
df['lead1'] = df['val'].shift(-1)   # lead 1 (next row)
df['diff1'] = df['val'].diff(1)     # first difference
df['pct']   = df['val'].pct_change()  # % change
```

## 25. CHAINING & ASSIGN PATTERN

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({'name':['Alice','Bob','Carol','Dave'],
                   'dept':['IT','HR','IT','HR'],
                   'salary':[100,200,150,np.nan]})

# Method chaining — clean, readable pipeline
result = (
    df
    .rename(columns=str.lower)
    .dropna(subset=['salary'])
    .assign(
        salary_k = lambda x: x['salary'] / 1000,
        is_it    = lambda x: x['dept'] == 'IT'
    )
    .query('salary > 100')
    .sort_values('salary', ascending=False)
    .reset_index(drop=True)
    [['name','dept','salary','salary_k']]
)

# pipe with external function
def log_shape(df, msg=''):
    print(f'{msg} shape: {df.shape}')
    return df

result = (
    df
    .pipe(log_shape, 'before')
    .dropna()
    .pipe(log_shape, 'after')
)
```

## 26. PERFORMANCE TIPS

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({'A': np.arange(1_000_000), 'B': np.random.rand(1_000_000)})

# 1. Vectorized ops over loops
df['C'] = df['A'] * df['B']         # fast
# [df['A'][i] * df['B'][i] for i in range(len(df))]  # slow

# 2. Avoid loc/iloc in loops → use vectorized + np.where
df['label'] = np.where(df['A'] > 500_000, 'high', 'low')

# 3. Use query for filtering (faster on large df)
df.query('A > 100 and B < 0.5')

# 4. categorical dtype — saves memory for low-cardinality
df_cat = pd.DataFrame({'dept': ['IT','HR'] * 500_000})
df_cat['dept'].nbytes                           # before
df_cat['dept'] = df_cat['dept'].astype('category')
df_cat['dept'].nbytes                           # after (much smaller)

# 5. float32 instead of float64
df['B'] = df['B'].astype(np.float32)

# 6. Read CSV with dtypes specified (skip auto-detection)
# pd.read_csv('data.csv', dtype={'A': np.int32, 'B': np.float32})

# 7. chunking for large files
# chunks = pd.read_csv('big.csv', chunksize=10_000)
# result = pd.concat([chunk[chunk['A'] > 0] for chunk in chunks])

# 8. eval for complex expressions (avoids temp arrays)
df.eval('D = A + B * 2', inplace=True)

# 9. copy() to avoid SettingWithCopyWarning
subset = df[df['A'] > 100].copy()
subset['E'] = 99     # safe — working on copy

# 10. Memory usage check
df.memory_usage(deep=True).sum() / 1024**2    # MB
```
