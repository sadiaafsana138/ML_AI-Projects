# 📊 Matplotlib Master Cheatsheet
> 100% coverage for ML / Data Science / Interview prep


```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.gridspec import GridSpec
from mpl_toolkits.mplot3d import Axes3D

np.random.seed(42)
x = np.arange(1, 11)
y = x ** 2
print('Matplotlib version:', mpl.__version__)
```

---
## 1. BASIC PLOTS

```python
# Line
plt.plot(x, y)
plt.title('Line Plot')
plt.xlabel('X'); plt.ylabel('Y')
plt.show()

# Scatter
plt.scatter(x, y)
plt.show()

# Bar (vertical)
plt.bar(x, y)
plt.show()

# Bar (horizontal)
plt.barh(x, y)
plt.show()

# Histogram
plt.hist(y, bins=5)
plt.show()

# Pie
plt.pie([10, 20, 30], labels=['A','B','C'], autopct='%1.1f%%')
plt.show()

# Step plot
plt.step(x, y, where='mid')
plt.show()

# Area / fill_between
plt.fill_between(x, y, alpha=0.3)
plt.show()

# Stem plot
plt.stem(x, y)
plt.show()

# Box plot
data = [np.random.randn(50) for _ in range(4)]
plt.boxplot(data, labels=['A','B','C','D'])
plt.show()

# Violin plot
plt.violinplot(data)
plt.show()

# Broken bar (Gantt-style)
plt.broken_barh([(10, 30), (55, 25)], (20, 9), facecolors='steelblue')
plt.show()
```

---
## 2. LINE PLOT OPTIONS

```python
# Colors
plt.plot(x, y, color='red')          # named
plt.plot(x, y, color='#FF5733')      # hex
plt.plot(x, y, color=(0.1,0.5,0.9)) # RGB tuple
plt.plot(x, y, color='C3')           # default color cycle

# Linestyles
plt.plot(x, y, linestyle='-')        # solid (default)
plt.plot(x, y, linestyle='--')       # dashed
plt.plot(x, y, linestyle='-.')       # dash-dot
plt.plot(x, y, linestyle=':')        # dotted
plt.plot(x, y, ls='None')            # no line (just markers)

# Markers
plt.plot(x, y, marker='o')           # circle
plt.plot(x, y, marker='s')           # square
plt.plot(x, y, marker='^')           # triangle up
plt.plot(x, y, marker='v')           # triangle down
plt.plot(x, y, marker='D')           # diamond
plt.plot(x, y, marker='*')           # star
plt.plot(x, y, marker='+')           # plus
plt.plot(x, y, marker='x')           # x
plt.plot(x, y, marker='|')           # vline

# Width & size
plt.plot(x, y, linewidth=2.5)
plt.plot(x, y, markersize=10, markerfacecolor='red', markeredgecolor='black')

# Alpha (transparency)
plt.plot(x, y, alpha=0.5)

# Short format string: [color][marker][linestyle]
plt.plot(x, y, 'r--o')              # red dashed with circle markers
plt.plot(x, y, 'bs-')               # blue square solid
plt.show()
```

---
## 3. SCATTER PLOT OPTIONS

```python
x = np.random.rand(50)
y = np.random.rand(50)
sizes  = np.random.rand(50) * 200
colors = np.random.rand(50)

# Basic
plt.scatter(x, y)

# Variable size & color (bubble chart)
sc = plt.scatter(x, y, s=sizes, c=colors, cmap='viridis', alpha=0.7)
plt.colorbar(sc, label='Value')    # add colorbar
plt.show()

# Edge color
plt.scatter(x, y, edgecolors='black', linewidths=0.5)

# Marker style
plt.scatter(x, y, marker='^', s=100)
plt.show()
```

---
## 4. BAR PLOT OPTIONS

```python
categories = ['IT','HR','Finance','Marketing']
values1 = [100, 200, 150, 180]
values2 = [120, 180, 160, 200]

# Basic
plt.bar(categories, values1)
plt.show()

# Color per bar
plt.bar(categories, values1, color=['blue','red','green','orange'])
plt.show()

# Grouped bar chart
x_pos = np.arange(len(categories))
width = 0.35
plt.bar(x_pos - width/2, values1, width, label='2023')
plt.bar(x_pos + width/2, values2, width, label='2024')
plt.xticks(x_pos, categories)
plt.legend()
plt.show()

# Stacked bar
plt.bar(categories, values1, label='2023')
plt.bar(categories, values2, bottom=values1, label='2024')
plt.legend()
plt.show()

# Value labels on bars
bars = plt.bar(categories, values1)
for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2,
             bar.get_height() + 2,
             str(bar.get_height()),
             ha='center', va='bottom')
plt.show()
```

---
## 5. HISTOGRAM OPTIONS

```python
data = np.random.randn(1000)

# Basic
plt.hist(data, bins=30)
plt.show()

# Density (normalize to probability)
plt.hist(data, bins=30, density=True, alpha=0.7)
plt.show()

# Cumulative
plt.hist(data, bins=30, cumulative=True)
plt.show()

# Overlapping histograms
data2 = np.random.randn(1000) + 2
plt.hist(data, bins=30, alpha=0.5, label='Group A')
plt.hist(data2, bins=30, alpha=0.5, label='Group B')
plt.legend()
plt.show()

# Horizontal histogram
plt.hist(data, bins=30, orientation='horizontal')
plt.show()

# 2D histogram (heatmap)
plt.hist2d(data, data2, bins=30, cmap='Blues')
plt.colorbar()
plt.show()

# Hexbin
plt.hexbin(data, data2, gridsize=20, cmap='inferno')
plt.colorbar()
plt.show()
```

## 6. PIE CHART OPTIONS

```python
sizes  = [30, 25, 20, 15, 10]
labels = ['IT','HR','Finance','Marketing','Ops']
colors = ['steelblue','coral','gold','lightgreen','orchid']
explode = [0.1, 0, 0, 0, 0]   # explode first slice

plt.pie(sizes,
        labels=labels,
        colors=colors,
        explode=explode,
        autopct='%1.1f%%',
        shadow=True,
        startangle=90)

plt.axis('equal')  # keep circular
plt.title('Department Distribution')
plt.show()

# Donut chart (pie with white center)
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, wedgeprops=dict(width=0.5))
plt.show()
```

## 7. FIGURE & AXES — OOP STYLE (RECOMMENDED)

```python
x = np.arange(1, 11)
y = x ** 2

# Create figure + axes
fig, ax = plt.subplots(figsize=(8, 5))   # single plot

ax.plot(x, y, 'r--o', label='x^2')
ax.set_title('OOP Style Plot', fontsize=14, fontweight='bold')
ax.set_xlabel('X Axis', fontsize=12)
ax.set_ylabel('Y Axis', fontsize=12)
ax.set_xlim(0, 11)
ax.set_ylim(0, 110)
ax.set_xticks(x)
ax.set_xticklabels([f'val_{i}' for i in x], rotation=45)
ax.legend(loc='upper left')
ax.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

# Figure properties
fig.set_size_inches(10, 6)
fig.set_dpi(100)
fig.suptitle('Figure Title', fontsize=16, y=1.02)  # overall title
```

## 8. SUBPLOTS — ALL PATTERNS

```python
x = np.arange(1, 11)
y = x ** 2

# Basic grid
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
axes[0,0].plot(x, y)
axes[0,1].scatter(x, y)
axes[1,0].bar(x, y)
axes[1,1].hist(y)
plt.tight_layout()
plt.show()

# Flatten for easy iteration
fig, axes = plt.subplots(2, 3, figsize=(14, 8))
for ax in axes.flat:
    ax.plot(x, y)
    ax.set_title('subplot')
plt.tight_layout()
plt.show()

# Shared axes
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(10, 4))
ax1.plot(x, y)
ax2.plot(x, x*3)
plt.show()

# GridSpec — custom sized subplots
fig = plt.figure(figsize=(12, 8))
gs  = GridSpec(2, 3, figure=fig, hspace=0.4, wspace=0.3)

ax1 = fig.add_subplot(gs[0, :])     # top row, full width
ax2 = fig.add_subplot(gs[1, 0])     # bottom left
ax3 = fig.add_subplot(gs[1, 1:])    # bottom right (spans 2 cols)

ax1.plot(x, y);    ax1.set_title('Wide')
ax2.bar(x[:5], y[:5]);  ax2.set_title('Bar')
ax3.scatter(x, y); ax3.set_title('Scatter')
plt.show()

# subplot2grid
fig = plt.figure(figsize=(10, 6))
ax1 = plt.subplot2grid((3,3), (0,0), colspan=3)
ax2 = plt.subplot2grid((3,3), (1,0), colspan=2, rowspan=2)
ax3 = plt.subplot2grid((3,3), (1,2), rowspan=2)
ax1.plot(x, y);   ax1.set_title('Top Full')
ax2.scatter(x,y); ax2.set_title('Left')
ax3.bar(x[:5], y[:5]); ax3.set_title('Right')
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(x, y)

# Grid
ax.grid(True)
ax.grid(True, axis='x')                      # only x-axis grid
ax.grid(True, linestyle='--', color='gray', alpha=0.5, linewidth=0.8)
ax.grid(True, which='both')                  # major + minor grid
ax.grid(True, which='minor', alpha=0.2)

# Spines (the border lines)
ax.spines['top'].set_visible(False)          # hide top border
ax.spines['right'].set_visible(False)        # hide right border
ax.spines['left'].set_color('red')
ax.spines['bottom'].set_linewidth(2)

# Move spines
ax.spines['left'].set_position('zero')       # move to x=0
ax.spines['bottom'].set_position('zero')     # move to y=0

plt.show()
```

## 11. ANNOTATION & TEXT

```python
fig, ax = plt.subplots(figsize=(8,5))
ax.plot(x, y)

# Simple text
ax.text(5, 25, 'Some text', fontsize=12, color='red')
ax.text(3, 70, 'Bold text', fontweight='bold', ha='center')

# Arrow annotation
ax.annotate('Peak here',
            xy=(10, 100),           # point to
            xytext=(7, 80),         # text at
            fontsize=10,
            arrowprops=dict(facecolor='black', arrowstyle='->', lw=1.5))

# Annotate with box
ax.annotate('Boxed',
            xy=(5, 25), xytext=(3, 50),
            bbox=dict(boxstyle='round,pad=0.3', facecolor='yellow', alpha=0.5),
            arrowprops=dict(arrowstyle='->'))

# Math text (LaTeX)
ax.text(6, 40, r'$y = x^2$', fontsize=14)

# axhline / axvline — reference lines
ax.axhline(y=50, color='gray', linestyle='--', linewidth=1, label='y=50')
ax.axvline(x=5,  color='blue', linestyle=':',  linewidth=1, label='x=5')

# axhspan / axvspan — shaded band
ax.axhspan(20, 60, alpha=0.1, color='green')
ax.axvspan(3, 7,   alpha=0.1, color='red')

plt.show()
```

## 12. SCALES & LIMITS

```python
x = np.arange(1, 11)
y = x ** 2

# Limits
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_xlim(0, 12)
ax.set_ylim(-10, 120)
ax.set_xlim(left=0)      # set only one side

# Log scale
ax.set_yscale('log')
ax.set_xscale('log')
ax.set_yscale('symlog')    # handles zero/negatives
ax.set_yscale('logit')     # for [0,1] probabilities
plt.show()

# Invert axis
fig, ax = plt.subplots()
ax.plot(x, y)
ax.invert_xaxis()
ax.invert_yaxis()
plt.show()

# Twin axis (secondary y-axis)
fig, ax1 = plt.subplots(figsize=(8,5))
ax2 = ax1.twinx()
ax1.plot(x, y,    'b-', label='x^2')
ax2.plot(x, x*3,  'r--', label='3x')
ax1.set_ylabel('y = x^2', color='blue')
ax2.set_ylabel('y = 3x',  color='red')
fig.legend(loc='upper left', bbox_to_anchor=(0.1,0.9))
plt.show()

# Twin x-axis
ax3 = ax1.twiny()
```

## 16. CONTOUR PLOTS

```python
xs = np.linspace(-3, 3, 100)
ys = np.linspace(-3, 3, 100)
XX, YY = np.meshgrid(xs, ys)
ZZ = np.sin(XX) * np.cos(YY)

# Contour lines
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

c1 = axes[0].contour(XX, YY, ZZ, levels=15, cmap='RdBu')
axes[0].clabel(c1, inline=True, fontsize=8)   # label lines
axes[0].set_title('Contour Lines')

c2 = axes[1].contourf(XX, YY, ZZ, levels=15, cmap='viridis')  # filled
plt.colorbar(c2, ax=axes[1])
axes[1].set_title('Filled Contour')

axes[2].contourf(XX, YY, ZZ, levels=15, cmap='viridis')
axes[2].contour(XX, YY, ZZ, levels=15, colors='black', linewidths=0.5)
axes[2].set_title('Combined')

plt.tight_layout()
plt.show()
```

## 17. ERROR BARS

```python
x = np.arange(1, 6)
y = np.array([2, 4, 3, 5, 4])
yerr = np.array([0.3, 0.5, 0.4, 0.6, 0.3])
xerr = np.array([0.1, 0.2, 0.1, 0.2, 0.1])

# Y error only
plt.errorbar(x, y, yerr=yerr, fmt='o-', capsize=5, label='y error')

# Both X and Y errors
plt.errorbar(x, y+1, xerr=xerr, yerr=yerr,
             fmt='s--', capsize=5, capthick=2,
             elinewidth=1.5, ecolor='red', label='xy error')

# Asymmetric errors
err_low  = np.array([0.2, 0.3, 0.1, 0.4, 0.2])
err_high = np.array([0.5, 0.4, 0.3, 0.6, 0.3])
plt.errorbar(x, y-1, yerr=[err_low, err_high], fmt='D:', capsize=5, label='asymmetric')

plt.legend()
plt.show()
```

## 18. FILL BETWEEN & BANDS

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Fill between two curves
plt.plot(x, y1, label='sin')
plt.plot(x, y2, label='cos')
plt.fill_between(x, y1, y2, alpha=0.2, color='purple', label='area')
plt.legend()
plt.show()

# Fill based on condition
plt.fill_between(x, y1, 0,
    where=(y1 > 0), color='green', alpha=0.3, label='positive')
plt.fill_between(x, y1, 0,
    where=(y1 < 0), color='red',   alpha=0.3, label='negative')
plt.plot(x, y1, 'k-')
plt.legend()
plt.show()

# Confidence interval style
mean = np.sin(x)
std  = 0.2
plt.plot(x, mean, 'b-', label='mean')
plt.fill_between(x, mean - std, mean + std,
    alpha=0.2, color='blue', label='±1 std')
plt.fill_between(x, mean - 2*std, mean + 2*std,
    alpha=0.1, color='blue', label='±2 std')
plt.legend()
plt.show()
```

## 19. 3D PLOTS

```python
from mpl_toolkits.mplot3d import Axes3D

# 3D Scatter
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')
X = np.random.rand(50); Y = np.random.rand(50); Z = np.random.rand(50)
ax.scatter(X, Y, Z, c=Z, cmap='viridis', s=50)
ax.set_xlabel('X'); ax.set_ylabel('Y'); ax.set_zlabel('Z')
plt.show()

# 3D Line
t = np.linspace(0, 4*np.pi, 200)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(np.sin(t), np.cos(t), t, lw=1.5)
plt.show()

# 3D Surface
xs = np.linspace(-3, 3, 50)
ys = np.linspace(-3, 3, 50)
XX, YY = np.meshgrid(xs, ys)
ZZ = np.sin(np.sqrt(XX**2 + YY**2))

fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(XX, YY, ZZ, cmap='viridis', alpha=0.8)
fig.colorbar(surf, shrink=0.5)
plt.show()

# 3D Wireframe
fig = plt.figure()
ax  = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(XX, YY, ZZ, linewidth=0.5)
plt.show()

# 3D Bar
fig = plt.figure()
ax  = fig.add_subplot(111, projection='3d')
xp = [1, 2, 3]; yp = [1, 2, 3]; zp = [0, 0, 0]
dz = [10, 20, 15]
ax.bar3d(xp, yp, zp, dx=0.5, dy=0.5, dz=dz)
plt.show()

# View angle
ax.view_init(elev=30, azim=45)   # elevation and azimuth
```

## 22. ANIMATION

```python
from matplotlib.animation import FuncAnimation

fig, ax = plt.subplots(figsize=(6,4))
ax.set_xlim(0, 2*np.pi)
ax.set_ylim(-1.5, 1.5)
line, = ax.plot([], [], 'b-', lw=2)

def init():
    line.set_data([], [])
    return line,

def update(frame):
    t = np.linspace(0, 2*np.pi, 100)
    y = np.sin(t + frame / 10)
    line.set_data(t, y)
    return line,

ani = FuncAnimation(fig, update, frames=100,
                    init_func=init, blit=True, interval=50)

# Display in notebook
from IPython.display import HTML
HTML(ani.to_jshtml())

# Save animation
# ani.save('animation.gif', writer='pillow', fps=20)
# ani.save('animation.mp4', writer='ffmpeg', fps=20)
plt.show()
```
