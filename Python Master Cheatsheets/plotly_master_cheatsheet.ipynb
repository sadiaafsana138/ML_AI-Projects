```python
import plotly.express as px
import plotly.graph_objects as go
import plotly.figure_factory as ff
from plotly.subplots import make_subplots
import plotly.io as pio
import pandas as pd
import numpy as np

np.random.seed(42)
print('Plotly version:', go.__class__.__module__.split('.')[0])

# ── Reusable sample data ────────────────────────────────────────
df = pd.DataFrame({
    'x':   np.arange(30),
    'y':   np.random.randint(1, 100, 30),
    'y2':  np.random.randint(1, 100, 30),
    'z':   np.random.randint(1, 100, 30),
    'cat': np.random.choice(['A','B','C'], 30),
    'grp': np.random.choice(['G1','G2'], 30),
    'val': np.random.randn(30) * 10 + 50
})

# Built-in datasets
tips     = px.data.tips()
iris     = px.data.iris()
gapminder= px.data.gapminder()
stocks   = px.data.stocks()
wind     = px.data.wind()
```

---
## 1. PLOTLY EXPRESS — LINE / SCATTER / AREA

```python
# ── Line ────────────────────────────────────────────────────────
fig = px.line(df, x='x', y='y', title='Line Plot')
fig.show()

# Multiple lines via hue
fig = px.line(stocks, x='date', y=['GOOG','AAPL','AMZN'],
              title='Stock Prices')
fig.show()

# Line with markers
fig = px.line(df, x='x', y='y', markers=True)
fig.show()

# ── Scatter ─────────────────────────────────────────────────────
fig = px.scatter(tips, x='total_bill', y='tip',
                 color='sex',        # color grouping
                 symbol='smoker',    # marker symbol
                 size='size',        # marker size
                 hover_data=['day'],  # extra tooltip info
                 title='Tips Scatter')
fig.show()

# Bubble chart
fig = px.scatter(gapminder.query('year==2007'),
                 x='gdpPercap', y='lifeExp',
                 size='pop', color='continent',
                 hover_name='country',
                 log_x=True, size_max=60)
fig.show()

# Scatter matrix
fig = px.scatter_matrix(iris, dimensions=['sepal_width','sepal_length',
                                           'petal_width','petal_length'],
                         color='species')
fig.show()

# ── Area ────────────────────────────────────────────────────────
fig = px.area(stocks, x='date', y='GOOG', title='Area Chart')
fig.show()

# Stacked area
fig = px.area(stocks, x='date', y=['GOOG','AAPL','AMZN'])
fig.show()
```

---
## 2. BAR PLOTS

```python
# Basic bar
fig = px.bar(tips, x='day', y='total_bill', title='Bar Chart')
fig.show()

# Grouped bar
fig = px.bar(tips, x='day', y='total_bill', color='sex',
             barmode='group')           # group | overlay | relative
fig.show()

# Stacked bar (default)
fig = px.bar(tips, x='day', y='total_bill', color='sex',
             barmode='stack')
fig.show()

# Horizontal bar
fig = px.bar(tips, x='total_bill', y='day',
             color='sex', orientation='h')
fig.show()

# Sorted bar
fig = px.bar(tips.groupby('day', as_index=False)['total_bill'].mean(),
             x='day', y='total_bill',
             category_orders={'day': ['Thur','Fri','Sat','Sun']})
fig.show()

# Text labels on bars
fig = px.bar(tips, x='day', y='total_bill', text_auto='.1f')
fig.update_traces(textposition='outside')
fig.show()
```

---
## 3. DISTRIBUTION PLOTS

```python
# ── Histogram ───────────────────────────────────────────────────
fig = px.histogram(tips, x='total_bill', nbins=20)
fig.show()

# With KDE
fig = px.histogram(tips, x='total_bill', marginal='violin',
                   hover_data=tips.columns)   # marginal: box, violin, rug
fig.show()

# Grouped
fig = px.histogram(tips, x='total_bill', color='sex',
                   barmode='overlay', opacity=0.7)
fig.show()

# Normalized
fig = px.histogram(tips, x='total_bill', color='sex',
                   histnorm='probability density')  # count, percent, density, probability
fig.show()

# Cumulative
fig = px.histogram(tips, x='total_bill', cumulative=True)
fig.show()

# ── Box ─────────────────────────────────────────────────────────
fig = px.box(tips, x='day', y='total_bill', color='sex',
             notched=True,           # notch for median CI
             points='all')           # all | outliers | False
fig.show()

# ── Violin ──────────────────────────────────────────────────────
fig = px.violin(tips, x='day', y='total_bill', color='sex',
                box=True,            # inner box
                points='all',
                violinmode='overlay')  # overlay | group
fig.show()

# ── Strip ───────────────────────────────────────────────────────
fig = px.strip(tips, x='day', y='total_bill', color='sex')
fig.show()

# ── ECDF ────────────────────────────────────────────────────────
fig = px.ecdf(tips, x='total_bill', color='sex')
fig.show()
```

---
## 4. CATEGORICAL & STATISTICAL PLOTS

```python
# ── Pie ─────────────────────────────────────────────────────────
fig = px.pie(tips, names='day', values='total_bill',
             title='Sales by Day', hole=0)   # hole=0.4 for donut
fig.show()

# Donut
fig = px.pie(tips, names='day', values='total_bill', hole=0.4)
fig.show()

# Pull a slice
fig = px.pie(tips, names='day', values='total_bill')
fig.update_traces(pull=[0.1, 0, 0, 0])     # explode first slice
fig.show()

# ── Funnel ──────────────────────────────────────────────────────
funnel_df = pd.DataFrame({'stage': ['Leads','Qualified','Proposal','Won'],
                           'count': [500, 300, 150, 50]})
fig = px.funnel(funnel_df, x='count', y='stage')
fig.show()

# ── Treemap ─────────────────────────────────────────────────────
fig = px.treemap(tips, path=['day','sex'], values='total_bill')
fig.show()

# ── Sunburst ────────────────────────────────────────────────────
fig = px.sunburst(tips, path=['day','time','sex'], values='total_bill')
fig.show()

# ── Parallel coordinates ─────────────────────────────────────────
fig = px.parallel_coordinates(iris, color='species_id',
                               dimensions=['sepal_length','sepal_width',
                                           'petal_length','petal_width'])
fig.show()

# ── Parallel categories ──────────────────────────────────────────
fig = px.parallel_categories(tips, dimensions=['day','time','sex'],
                              color='total_bill', color_continuous_scale='Blues')
fig.show()

# ── Icicle / packed circles ──────────────────────────────────────
fig = px.icicle(tips, path=['day','sex'], values='total_bill')
fig.show()
```

---
## 5. HEATMAP & MATRIX

```python
# px.imshow — correlation or any 2D array
corr = tips.select_dtypes('number').corr()
fig = px.imshow(corr, text_auto='.2f', color_continuous_scale='RdBu',
                zmin=-1, zmax=1, aspect='auto')
fig.show()

# Pivot heatmap
flights = px.data.stocks()   # use custom pivot instead
pivot = tips.pivot_table(index='day', columns='time',
                          values='total_bill', aggfunc='mean')
fig = px.imshow(pivot, text_auto='.1f', color_continuous_scale='Blues')
fig.show()

# go.Heatmap (more control)
z = np.random.rand(5, 5)
fig = go.Figure(go.Heatmap(
    z=z,
    x=['A','B','C','D','E'],
    y=['V','W','X','Y','Z'],
    colorscale='Viridis',
    text=np.round(z, 2),
    texttemplate='%{text}',
    showscale=True
))
fig.update_layout(title='Custom Heatmap')
fig.show()

# Annotated heatmap via figure_factory
z_data = np.round(np.random.rand(4,4), 2)
fig = ff.create_annotated_heatmap(z_data,
    x=['W1','W2','W3','W4'], y=['R1','R2','R3','R4'],
    colorscale='Blues')
fig.show()
```

## 6. MAP & GEO PLOTS

```python
# ── Choropleth (country/region fill) ────────────────────────────
gap07 = gapminder.query('year==2007')
fig = px.choropleth(gap07, locations='iso_alpha',
                    color='lifeExp',
                    hover_name='country',
                    color_continuous_scale='RdYlGn',
                    title='Life Expectancy 2007')
fig.show()

# Scope zoom
fig = px.choropleth(gap07, locations='iso_alpha', color='gdpPercap',
                    scope='europe', color_continuous_scale='Blues')
fig.show()

# ── Scatter geo ─────────────────────────────────────────────────
fig = px.scatter_geo(gap07, locations='iso_alpha',
                     size='pop', color='continent',
                     hover_name='country',
                     projection='natural earth')
fig.show()

# ── Mapbox scatter (needs lat/lon) ───────────────────────────────
# px.scatter_mapbox / px.density_mapbox / px.line_mapbox
# Requires mapbox token or use open-street-map style
cities = pd.DataFrame({
    'city': ['Dhaka','Chittagong','Rajshahi'],
    'lat':  [23.8103, 22.3569, 24.3745],
    'lon':  [90.4125, 91.7832, 88.6042],
    'pop':  [8906035, 2592439, 700000]
})
fig = px.scatter_mapbox(cities, lat='lat', lon='lon',
                         size='pop', hover_name='city',
                         mapbox_style='open-street-map', zoom=5)
fig.show()
```

## 7. 3D PLOTS

```python
# ── 3D Scatter ──────────────────────────────────────────────────
fig = px.scatter_3d(iris, x='sepal_length', y='sepal_width', z='petal_length',
                    color='species', symbol='species', size_max=8)
fig.show()

# ── 3D Line ─────────────────────────────────────────────────────
t = np.linspace(0, 4*np.pi, 200)
fig = px.line_3d(x=np.sin(t), y=np.cos(t), z=t)
fig.show()

# ── 3D Surface (go.Surface) ──────────────────────────────────────
xs = np.linspace(-3, 3, 50)
ys = np.linspace(-3, 3, 50)
XX, YY = np.meshgrid(xs, ys)
ZZ = np.sin(np.sqrt(XX**2 + YY**2))

fig = go.Figure(go.Surface(x=XX, y=YY, z=ZZ,
                            colorscale='Viridis',
                            contours_z=dict(show=True, usecolormap=True)))
fig.update_layout(title='3D Surface', scene=dict(
    xaxis_title='X', yaxis_title='Y', zaxis_title='Z'))
fig.show()

# ── 3D Mesh ──────────────────────────────────────────────────────
fig = go.Figure(go.Mesh3d(x=[0,1,2,0], y=[0,0,1,2],
                           z=[0,1,0,0], color='lightblue',
                           opacity=0.6))
fig.show()

# Camera / scene control
fig.update_layout(scene_camera=dict(
    eye=dict(x=1.5, y=1.5, z=0.8)))  # camera angle
```

## 8. GRAPH OBJECTS (go) — LOW LEVEL

```python
# go gives full manual control; px is built on top of go

# ── Scatter / Line modes ─────────────────────────────────────────
fig = go.Figure()
fig.add_trace(go.Scatter(
    x=df['x'], y=df['y'],
    mode='lines+markers',   # lines | markers | lines+markers | text
    name='Series A',
    line=dict(color='royalblue', width=2, dash='solid'),  # solid dash dot dashdot
    marker=dict(size=8, color='red', symbol='circle',     # circle square diamond x
                line=dict(color='black', width=1))
))
fig.show()

# ── Bar ─────────────────────────────────────────────────────────
fig = go.Figure([
    go.Bar(name='2023', x=['IT','HR','Finance'], y=[100,200,150]),
    go.Bar(name='2024', x=['IT','HR','Finance'], y=[120,180,160])
])
fig.update_layout(barmode='group')
fig.show()

# ── Histogram ────────────────────────────────────────────────────
fig = go.Figure(go.Histogram(
    x=tips['total_bill'], nbinsx=20,
    histnorm='probability density',
    marker_color='steelblue', opacity=0.8
))
fig.show()

# ── Box ─────────────────────────────────────────────────────────
fig = go.Figure(go.Box(
    y=tips['total_bill'], x=tips['day'],
    boxpoints='outliers',   # all | outliers | suspectedoutliers | False
    notched=True,
    marker_color='indianred'
))
fig.show()

# ── Pie ─────────────────────────────────────────────────────────
fig = go.Figure(go.Pie(
    labels=['A','B','C'], values=[30,50,20],
    hole=0.3,
    pull=[0.1,0,0],
    textinfo='label+percent'
))
fig.show()

# ── Waterfall ────────────────────────────────────────────────────
fig = go.Figure(go.Waterfall(
    name='P&L',
    orientation='v',
    x=['Revenue','COGS','Gross','OpEx','EBIT'],
    measure=['absolute','relative','total','relative','total'],
    y=[500,-200,0,-100,0]
))
fig.show()

# ── Indicator (KPI card) ─────────────────────────────────────────
fig = go.Figure(go.Indicator(
    mode='number+delta+gauge',
    value=92.1,
    delta={'reference': 85, 'valueformat': '.1f'},
    gauge={'axis': {'range': [0, 100]},
           'bar':  {'color': 'royalblue'},
           'threshold': {'line': {'color': 'red', 'width': 4},
                         'thickness': 0.75, 'value': 90}},
    title={'text': 'Accuracy %'}
))
fig.show()

# ── Candlestick ──────────────────────────────────────────────────
candle_df = pd.DataFrame({
    'date':  pd.date_range('2024-01-01', periods=30),
    'open':  np.random.randint(100,200,30).astype(float),
    'high':  np.random.randint(200,250,30).astype(float),
    'low':   np.random.randint(80, 100,30).astype(float),
    'close': np.random.randint(100,200,30).astype(float)
})
fig = go.Figure(go.Candlestick(
    x=candle_df['date'],
    open=candle_df['open'], high=candle_df['high'],
    low=candle_df['low'],   close=candle_df['close']
))
fig.show()
```

## 9. SUBPLOTS

```python
# ── Basic grid ───────────────────────────────────────────────────
fig = make_subplots(rows=2, cols=2,
                    subplot_titles=['Scatter','Bar','Histogram','Box'])

fig.add_trace(go.Scatter(x=df['x'], y=df['y'], mode='markers'), row=1, col=1)
fig.add_trace(go.Bar(x=df['x'][:6], y=df['y'][:6]),            row=1, col=2)
fig.add_trace(go.Histogram(x=df['y'], nbinsx=10),               row=2, col=1)
fig.add_trace(go.Box(y=df['y']),                                 row=2, col=2)

fig.update_layout(height=600, title_text='Dashboard')
fig.show()

# ── Shared axes ──────────────────────────────────────────────────
fig = make_subplots(rows=1, cols=2, shared_yaxes=True, shared_xaxes=False)
fig.add_trace(go.Scatter(x=df['x'], y=df['y'],  name='A'), row=1, col=1)
fig.add_trace(go.Scatter(x=df['x'], y=df['y2'], name='B'), row=1, col=2)
fig.show()

# ── Mixed types ──────────────────────────────────────────────────
fig = make_subplots(rows=1, cols=2,
                    specs=[[{'type':'xy'}, {'type':'pie'}]])
fig.add_trace(go.Bar(x=['A','B','C'], y=[10,20,30]), row=1, col=1)
fig.add_trace(go.Pie(labels=['A','B','C'], values=[10,20,30]), row=1, col=2)
fig.show()

# ── Spanning cells (colspan/rowspan) ─────────────────────────────
fig = make_subplots(
    rows=2, cols=3,
    specs=[
        [{'colspan': 2}, None, {'type':'pie'}],
        [{'colspan': 3}, None, None]
    ],
    subplot_titles=['Wide Top','Pie','Full Bottom']
)
fig.add_trace(go.Scatter(x=df['x'], y=df['y']), row=1, col=1)
fig.add_trace(go.Pie(labels=['A','B'], values=[60,40]),        row=1, col=3)
fig.add_trace(go.Bar(x=df['x'], y=df['y']),                   row=2, col=1)
fig.update_layout(height=500)
fig.show()

# ── Secondary y-axis ─────────────────────────────────────────────
fig = make_subplots(specs=[[{'secondary_y': True}]])
fig.add_trace(go.Scatter(x=df['x'], y=df['y'],  name='Sales'), secondary_y=False)
fig.add_trace(go.Scatter(x=df['x'], y=df['y2'], name='Rate'),  secondary_y=True)
fig.update_yaxes(title_text='Sales',    secondary_y=False)
fig.update_yaxes(title_text='Rate (%)', secondary_y=True)
fig.show()
```

## 10. update_layout — FULL CUSTOMIZATION

```python
fig = px.scatter(tips, x='total_bill', y='tip', color='sex')

fig.update_layout(
    # Title
    title=dict(text='Custom Layout', font=dict(size=20), x=0.5),

    # Size
    width=800, height=500,

    # Background
    paper_bgcolor='white',      # figure background
    plot_bgcolor='#f9f9f9',     # plot area background

    # Font
    font=dict(family='Arial', size=12, color='#333333'),

    # Margin
    margin=dict(l=60, r=30, t=80, b=60),

    # Legend
    legend=dict(
        title='Gender',
        x=0.01, y=0.99,
        bgcolor='rgba(255,255,255,0.8)',
        bordercolor='gray', borderwidth=1,
        orientation='v'    # h for horizontal
    ),
    showlegend=True,

    # Axes
    xaxis=dict(
        title='Total Bill ($)',
        range=[0, 55],
        showgrid=True, gridcolor='lightgray', gridwidth=0.5,
        zeroline=True, zerolinecolor='black',
        showline=True, linecolor='black',
        tickfont=dict(size=11),
        tickformat='$,.0f',
        dtick=10
    ),
    yaxis=dict(
        title='Tip ($)',
        showgrid=True, gridcolor='lightgray',
        tickformat='.1f'
    ),

    # Hover
    hovermode='closest',        # x | y | closest | x unified | False
)

fig.show()
```

## 12. ANNOTATIONS, SHAPES & IMAGES

```python
fig = px.line(df, x='x', y='y')

# ── Annotations (text + arrow) ───────────────────────────────────
fig.add_annotation(
    x=5, y=df['y'].iloc[5],
    text='Peak!',
    showarrow=True,
    arrowhead=2,
    ax=40, ay=-40,     # arrow offset
    font=dict(size=12, color='red'),
    bgcolor='lightyellow',
    bordercolor='gray'
)

# Multiple annotations at once
fig.update_layout(annotations=[
    dict(x=2, y=df['y'].iloc[2], text='Low',  showarrow=False),
    dict(x=8, y=df['y'].iloc[8], text='High', showarrow=False)
])

# ── Shapes ──────────────────────────────────────────────────────
# Horizontal line
fig.add_hline(y=df['y'].mean(), line_dash='dash',
              line_color='red', annotation_text='Mean')

# Vertical line
fig.add_vline(x=5, line_dash='dot', line_color='blue')

# Horizontal band
fig.add_hrect(y0=20, y1=60, fillcolor='lightgreen', opacity=0.2,
              line_width=0, annotation_text='Normal Range')

# Vertical band
fig.add_vrect(x0=3, x1=7, fillcolor='lightyellow', opacity=0.3, line_width=0)

# Custom shape
fig.update_layout(shapes=[
    dict(type='rect', x0=1, x1=4, y0=10, y1=80,
         fillcolor='lightblue', opacity=0.2, line_width=0),
    dict(type='circle', x0=7, x1=9, y0=20, y1=80,
         fillcolor='pink', opacity=0.3)
])

fig.show()
```

## 13. ANIMATION

```python
# ── px animation ─────────────────────────────────────────────────
fig = px.scatter(gapminder, x='gdpPercap', y='lifeExp',
                 animation_frame='year',
                 animation_group='country',
                 size='pop', color='continent',
                 hover_name='country',
                 log_x=True, size_max=55,
                 range_x=[100,100000], range_y=[25,90],
                 title='Gapminder Animation')
fig.show()

# Bar animation
fig = px.bar(gapminder, x='continent', y='pop',
             animation_frame='year',
             animation_group='continent',
             range_y=[0, 4e9])
fig.show()

# ── go frames (manual animation) ─────────────────────────────────
frames = []
for k in range(10):
    frames.append(go.Frame(
        data=[go.Scatter(x=np.arange(20),
                         y=np.sin(np.arange(20) + k*0.5))],
        name=str(k)
    ))

fig = go.Figure(
    data=[go.Scatter(x=np.arange(20), y=np.sin(np.arange(20)))],
    frames=frames,
    layout=go.Layout(
        updatemenus=[dict(type='buttons', showactive=False, buttons=[
            dict(label='Play', method='animate',
                 args=[None, {'frame': {'duration': 100, 'redraw': True},
                              'fromcurrent': True}]),
            dict(label='Pause', method='animate',
                 args=[[None], {'frame': {'duration': 0}, 'mode': 'immediate'}])
        ])]
    )
)
fig.show()
```
