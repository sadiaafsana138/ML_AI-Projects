## 📦 Imports

```python
import streamlit as st
import pandas as pd
import numpy as np
import time
import plotly.express as px
import matplotlib.pyplot as plt
import altair as alt
from io import BytesIO
```

## 1. PAGE CONFIG (Always First)

```python
# Must be the very first Streamlit call in your script
st.set_page_config(
    page_title='My App',
    page_icon='🗺️',             # emoji or path to .ico
    layout='wide',              # 'centered' or 'wide'
    initial_sidebar_state='expanded'  # 'expanded' or 'collapsed'
)
```

---
## 2. BASIC TEXT

```python
st.title('My App')
st.header('Header')
st.subheader('Subheader')
st.text('Plain monospace text')
st.write('Auto display — handles text, df, chart, dict, anything')
st.markdown('**Bold**, *italic*, `code`, [link](https://streamlit.io)')
st.caption('Small grey caption text')
st.code('print("hello")', language='python')
st.latex(r'E = mc^2')
st.divider()   # horizontal rule
```

---
## 3. DATA DISPLAY

```python
df = pd.DataFrame({
    'A': np.random.randint(1, 100, 10),
    'B': np.random.randint(1, 100, 10)
})

st.dataframe(df)                        # interactive, sortable
st.table(df)                            # static table
st.dataframe(df.style.highlight_max(axis=0))  # with styling
st.json({'key': 'value', 'num': 42})    # pretty JSON
st.write(df.describe())                 # summary stats
```

## 4. METRICS

```python
col1, col2, col3 = st.columns(3)

col1.metric('Sales',    '1,000', '+5%')
col2.metric('Revenue',  '$50K',  '-2%',  delta_color='inverse')  # red=up, green=down
col3.metric('Users',    '320',   '+12')
```

## 5. INPUT WIDGETS

## 19. PROGRESS & SPINNER

```python
# Spinner
with st.spinner('Loading...'):
    time.sleep(1)
st.success('Done!')

# Progress bar
bar = st.progress(0, text='Starting...')
for i in range(100):
    time.sleep(0.01)
    bar.progress(i + 1, text=f'Processing {i+1}%')
bar.empty()

# Status container (new in 1.28)
with st.status('Running pipeline...', expanded=True) as status:
    st.write('Step 1: Loading data')
    time.sleep(0.5)
    st.write('Step 2: Processing')
    time.sleep(0.5)
    status.update(label='Complete!', state='complete', expanded=False)
```

## 20. SESSION STATE

```python
# Persists values across reruns
if 'count' not in st.session_state:
    st.session_state.count = 0

if st.button('Increment'):
    st.session_state.count += 1

if st.button('Reset'):
    st.session_state.count = 0

st.write(f'Count: {st.session_state.count}')

# Store any type — list, dict, df, model
if 'history' not in st.session_state:
    st.session_state.history = []

if st.button('Add entry'):
    st.session_state.history.append(len(st.session_state.history))

st.write(st.session_state.history)
```

---
## 21. CACHING

```python
# @st.cache_data — for DataFrames, JSON, plain data
@st.cache_data
def load_data(path):
    return pd.read_csv(path)

# @st.cache_resource — for DB connections, ML models (shared across sessions)
@st.cache_resource
def load_model():
    # return joblib.load('model.pkl')
    return {'model': 'loaded'}

# TTL — auto-expire cache after N seconds
@st.cache_data(ttl=600)
def fetch_live_data():
    return pd.DataFrame(np.random.rand(10, 3))

# Clear cache manually
# load_data.clear()

data = fetch_live_data()
st.write(data)
```

## 22. FORMS (Batch Inputs, Single Submit)

```python
# Groups widgets — only triggers rerun on submit, not on every keystroke
with st.form('my_form'):
    st.subheader('User Details')
    name  = st.text_input('Name')
    email = st.text_input('Email')
    age   = st.number_input('Age', 0, 120)
    submitted = st.form_submit_button('Submit', type='primary')

if submitted:
    st.success(f'Submitted: {name}, {email}, {age}')
```

---
## 23. CALLBACKS (on_change / on_click)

```python
def on_change_handler():
    st.session_state.last_value = st.session_state.my_slider

if 'last_value' not in st.session_state:
    st.session_state.last_value = 0

st.slider(
    'Slider with callback',
    0, 100, 50,
    key='my_slider',
    on_change=on_change_handler
)

st.write(f'Last committed value: {st.session_state.last_value}')
```

## 25. CHAT INTERFACE

```python
# Built-in chat UI components
if 'messages' not in st.session_state:
    st.session_state.messages = []

# Render history
for msg in st.session_state.messages:
    with st.chat_message(msg['role']):   # role: 'user' or 'assistant'
        st.write(msg['content'])

# Input box pinned to bottom
if prompt := st.chat_input('Type a message...'):
    st.session_state.messages.append({'role': 'user', 'content': prompt})

    with st.chat_message('assistant'):
        response = f'Echo: {prompt}'
        st.write_stream(iter(response))  # streaming effect
        st.session_state.messages.append({'role': 'assistant', 'content': response})
```
